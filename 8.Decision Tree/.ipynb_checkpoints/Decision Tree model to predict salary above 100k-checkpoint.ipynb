```python
import pandas as pd
from sklearn.tree import tree
```

```python
default=pd.read_csv('salaries.csv')
```

```python
default.head()
```

```python
default
```

```python
#independent Variables
x=default.drop(['salary_more_then_100k'],axis='columns')
```

```python
x
```

```python
#dependent Variable/Target Varaible
y=default[['salary_more_then_100k']]
```

```python
model=tree.DecisionTreeClassifier()
```

```python
#Label Encoding 
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
default['CompanyLabel']=le_company.fit_transform(default['company'])
```

```python
default['JobLabel']=le_company.fit_transform(default['job'])
```

```python
default['DegreeLabel']=le_company.fit_transform(default['degree'])
```

```python
default
```

```python
decision_df=default.drop(['company','job','degree','salary_more_then_100k'],axis="columns")
```

```python
decision_target=default['salary_more_then_100k']
```

```python
model.fit(decision_df,decision_target)
```

### Predict wheater a person work in facebook on post of sales executive holds master degree earn more than 100k salary

```python
model.predict([[1,1,1]])
```

### Predict wheater a person work in Google on post of sales executive holds bachlore degree earn more than 100k salary

```python
model.predict([[0,2,2]])
```
